```python
# importing all packages for making model

import pandas as pd
import numpy as np
```

```python
# reading all data frame and getting relevent information

ratings = pd.read_csv('Ratings.csv')
user = pd.read_csv('Users.csv')
books = pd.read_csv('Books.csv')
```

```python
df = ratings.merge(books, on='ISBN')[['User-ID','Book-Title','Book-Rating','Image-URL-M']]
```

```python
df.head()
```

```python
# Number of rating got for a particular book
no_book_rating = df.groupby('Book-Title')['User-ID'].count().reset_index()
no_book_rating.rename(columns={'User-ID':'no_of_rating'},inplace=True)
no_book_rating.head()
```

```python
# Getting average rating per book

book_rating = df.groupby('Book-Title')['Book-Rating'].mean().reset_index()
book_rating.head()
```

```python
book_details = no_book_rating.merge(book_rating , on ='Book-Title')
book_details
```

```python
# Getting count of review made by every user 
number_of_rating = df.groupby('User-ID')['Book-Title'].count().reset_index()
number_of_rating.rename(columns={'Book-Title':'Count_of_rating'},inplace=True)
number_of_rating.head()
```

```python
number_of_rating.sort_values(by='Count_of_rating',ascending=False)
```

```python
# Getting list of all active user(i.e count of review greater than 10)
active_user = number_of_rating[number_of_rating['Count_of_rating']>10]['User-ID'].unique()
len(active_user)
```

```python
# Getting details of popular books 
popular_book = book_details[book_details['no_of_rating']>100]['Book-Title']
popular_book
```

```python
df.head()
```

```python
df = df[df['User-ID'].isin(active_user) & df['Book-Title'].isin(popular_book)]
```

```python
# creating pivot table for making vector representation of our data to feed that to a model
table = df.pivot_table(columns='User-ID',index='Book-Title',values='Book-Rating')
table
```

```python
# Filling NaN value with 0
table = table.fillna(0)
table
```

```python
pip install scikit-learn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Finding cosine similarity of vectors for calculating distance
similarity_score = cosine_similarity(table)
```

```python
sorted(list(enumerate(similarity_score[2])),key= lambda x:x[1],reverse=True)[1:11]
```

```python
# Creating function to get top 10 recommended books based on closest distance for a vector
def recommend(book_name):
    index = np.where(table.index == book_name)[0][0]
    data = sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1],reverse=True)[1:11]
    for i in data:
        print(table.index[i[0]])
```

```python
# Sample result of model
recommend('1984')
```
